```python
import numpy as np
from sklearn import datasets, metrics, linear_model
from sklearn import model_selection as modsel
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# ボストン住宅価格データセット
boston = datasets.load_boston()
```

```python
dir(boston)
```

```python
boston.data.shape # 合計506点のデータ点が含まれ、それぞれが13の特徴量を持つ
```

```python
# scikit-learnの線形回帰モデル
linreg = linear_model.LinearRegression()
```

```python
x_train, x_test, y_train, y_test = modsel.train_test_split(boston.data, boston.target,
                                                           test_size=0.1,
                                                           random_state=42
                                                          )
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
linreg.fit(x_train, y_train)
```

```python
# 平均二乗誤差
metrics.mean_squared_error(y_train, linreg.predict(x_train))
```

```python
# 決定係数
linreg.score(x_train, y_train)
```

### モデルの評価

```python
y_pred = linreg.predict(x_test)
metrics.mean_squared_error(y_test, y_pred)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, linewidth=2, label='ground truth')
plt.plot(y_pred, linewidth=2, label='predicted')
plt.legend(loc='best')
plt.xlabel('test data points')
plt.ylabel('target value')
plt.show()
```

```python
scorestr = 'R$^2$ = %.3f' % linreg.score(x_test, y_test)
errstr = 'MSE = %.3f' % metrics.mean_squared_error(y_test, y_pred)
plt.text(-5, 50, scorestr, fontsize=12)
plt.text(-5, 45, errstr, fontsize=12)
plt.plot(y_test, y_pred, 'o')
plt.plot([-10, 60], [-10, 60], 'k--')
plt.axis([-10, 60, -10, 60])
plt.xlabel('ground truth')
plt.ylabel('predicted')
plt.show()
```

```python
# ラッソ回帰アルゴリズム
lassoreg = linear_model.Lasso()

# リッジ回帰アルゴリズム
ridgereg = linear_model.Ridge()
```

```python
lassoreg.fit(x_train, y_train)
```

```python
y_pred = lassoreg.predict(x_test)
plt.figure(figsize=(10, 6))
plt.plot(y_test, linewidth=2, label='ground truth')
plt.plot(y_pred, linewidth=2, label='predicted')
plt.legend(loc='best')
plt.xlabel('test data points')
plt.ylabel('target value')
plt.show()
```

```python
scorestr = 'R$^2$ = %.3f' % lassoreg.score(x_test, y_test)
errstr = 'MSE = %.3f' % metrics.mean_squared_error(y_test, y_pred)
plt.text(-5, 50, scorestr, fontsize=12)
plt.text(-5, 45, errstr, fontsize=12)
plt.plot(y_test, y_pred, 'o')
plt.plot([-10, 60], [-10, 60], 'k--')
plt.axis([-10, 60, -10, 60])
plt.xlabel('ground truth')
plt.ylabel('predicted')
plt.show()
```

```python
ridgereg.fit(x_train, y_train)
y_pred = ridgereg.predict(x_test)
plt.figure(figsize=(10, 6))
plt.plot(y_test, linewidth=2, label='ground truth')
plt.plot(y_pred, linewidth=2, label='predicted')
plt.legend(loc='best')
plt.xlabel('test data points')
plt.ylabel('target value')
plt.show()

scorestr = 'R$^2$ = %.3f' % ridgereg.score(x_test, y_test)
errstr = 'MSE = %.3f' % metrics.mean_squared_error(y_test, y_pred)
plt.text(-5, 50, scorestr, fontsize=12)
plt.text(-5, 45, errstr, fontsize=12)
plt.plot(y_test, y_pred, 'o')
plt.plot([-10, 60], [-10, 60], 'k--')
plt.axis([-10, 60, -10, 60])
plt.xlabel('ground truth')
plt.ylabel('predicted')
plt.show()
```
