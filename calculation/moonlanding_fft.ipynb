```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2

%matplotlib inline
```

```python
img = Image.open("./Images/moonlanding.png")
plt.imshow(img, cmap="gray")
plt.show()
```

```python
gray_img = img.convert('L')
plt.imshow(gray_img, cmap="gray")
plt.show()
```

```python
np_img = np.asarray(gray_img)
np_img.shape
```

```python
ffted_img = np.fft.fft2(np_img)
shifted_f_uv = np.fft.fftshift(ffted_img)
unshifted_f_uv = np.fft.fftshift(shifted_f_uv)
i_f_xy = np.fft.ifft2(unshifted_f_uv).real
plt.imshow(i_f_xy, cmap="gray")
plt.show()
```

```python
magnitude_spectrum2d = 20 * np.log(np.absolute(shifted_f_uv))
magnitude_spectrum2d_var = 100000000000000000 * np.log(np.absolute(shifted_f_uv))

# figsize(6.4, 4.8)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 9))
ax = ax.flatten()
ax[0].imshow(magnitude_spectrum2d)
ax[1].imshow(magnitude_spectrum2d, cmap="gray")
ax[2].imshow(magnitude_spectrum2d_var)
ax[3].imshow(magnitude_spectrum2d_var, cmap="gray")
plt.show()
```
