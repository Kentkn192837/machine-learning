```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## データの作成

```python
width = height = 360
n = 6    # 描画する点数

np.random.seed(0)
pts = np.random.randint(0, width, (n, 2))    # 0以上width未満のn x 2の配列の整数の一様乱数
pts_0 = np.random.randint(0, width, (6, 2))

from scipy.spatial import Voronoi, voronoi_plot_2d

_, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
axes = axes.flatten()
vor = Voronoi(pts)  # Figureオブジェクト
vor_0 = Voronoi(pts_0)
ax0 = voronoi_plot_2d(vor, ax=axes[0])
ax1 = voronoi_plot_2d(vor_0, ax=axes[1])
plt.show()
```
