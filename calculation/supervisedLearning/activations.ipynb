```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## シグモイド関数

```python
x = np.linspace(-10, 10, 100)    # -10～10の間で100等分の値を出力する
y = 1 / (1 + np.exp(-x))

plt.plot(x, y, color='#0000ff')
plt.xticks(np.arange(-10, 12, 2))  # x軸の目盛りを設定
plt.yticks(np.arange(-3, 3, 0.5))  # y軸の目盛りを設定
plt.xlim(-10, 10)
plt.ylim(-1, 2)
plt.grid()
plt.show()
```

## ReLU(Rectified Linear Unit)

```python
def relu(X):
    y = np.array([])
    for x in X:
        if x <= 0:
            y = np.append(y, 0)
            continue
        y = np.append(y, x)
    return y

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.plot(x, y, color='#0000ff')
plt.xticks(np.arange(-10, 12, 2))
plt.yticks(np.arange(-2, 12, 1))
plt.xlim(-10, 10)
plt.ylim(-1, 10)
plt.grid()
plt.show()
```

## ハイパボリックタンジェント

```python
x = np.linspace(-10, 10, 100)
y = np.tanh(x)

plt.plot(x, y, color='#0000ff')
plt.xticks(np.arange(-10, 12, 2))
plt.yticks(np.arange(-2, 12, 0.5))
plt.xlim(-10, 10)
plt.ylim(-2, 2)
plt.grid()
plt.show()
```

## ソフトマックス

```python
def softmax(x0, x1, x2):
    u = np.exp(x0) + np.exp(x1) + np.exp(x2)
    return np.exp(x0) / u, np.exp(x1) / u, np.exp(x2) / u

y = softmax(3, 1, -2)
print(np.round(y, 2))         # 小数点以下2桁の概数を表示
print(np.sum(y))              # 和
print("--------------------")

y = softmax(2, 0, -1)
print(np.round(y, 2))         # 小数点以下2桁の概数を表示
print(np.sum(y))
```
